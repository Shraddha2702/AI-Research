```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV,StratifiedKFold
from sklearn.metrics import confusion_matrix, roc_auc_score ,roc_curve,auc
import warnings
warnings.filterwarnings("ignore")
```

## Data Download and Read Section

*All these is part of the Pre-processing*

```python
#Reading the File
lines = []
with open('spambase.data', 'r') as f:
    ls = f.readlines()
    for each in ls:
        lines.append(each.split(','))
```

```python
#Convert the String values to float values 
#so they can be used for the Models
data = []
for each in lines:
    chan = []
    for x in each:
        chan.append(float(x))
    data.append(chan)
```

```python
len(data[0]), len(data[1]), len(data[2])
```

Classification Problem   
58 Total Attributes - Multivariate / Real - valued   
Last column denotes whether the e-mail was considered spam (1) or not (0), i.e. unsolicited commercial e-mail. 

```python
#To get the column names
with open('spambase.names', 'r') as f:
    lls = f.readlines()
    
act = lls[33:]

headers = []
for each in act:
    headers.append(each.split(":")[0])
headers.append("spam")
```

```python
#Converting to pandas DF
df = pd.DataFrame(data, columns=headers)
```

### Useful Functions

```python
"""
Useful to find the Mean / Average of values
"""
def mean(data):
    summ = 0
    for each in data:
        summ += each
    return summ / len(data)

"""
Create a table with one row per fold showing your false positive, false negative, and overall error
rates, and add one final row corresponding to the average error rates across all folds.
"""
def display_results(tn, fp, fn, tp):
    # Fall out or false positive rate
    FPR = fp/(fp+tn)
    # False negative rate
    FNR = fn/(tp+fn)

    # Overall error
    ERROR = (fp+fn)/(tp+fp+fn+tn)
    return (FPR, FNR, ERROR)
```

## Building a Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#Splitting the Dataset into Train and Test Set
X = df.drop('spam',axis=1)
y = df['spam']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
all_classifier = []
```

*Tried various classifiers here. To select the best for the dataset.*

### Logisitic Regression

```python
from sklearn.linear_model import LogisticRegression

kf = StratifiedKFold(n_splits=5,shuffle=True,random_state=1)

cv_score =[]
all_vals = []
errors = []

i=1

for train_index,test_index in kf.split(X_train,y_train):
    print('{} of KFold {}'.format(i,kf.n_splits))
    xtr,xvl = X.loc[train_index],X.loc[test_index]
    ytr,yvl = y.loc[train_index],y.loc[test_index]
    
    #model
    lr = LogisticRegression(C=2).fit(xtr,ytr)
    tn, fp, fn, tp = confusion_matrix(yvl,lr.predict(xvl)).ravel()
    (FPR, FNR, ERROR) = display_results(tn, fp, fn, tp)
    all_vals.append([i, fp, fn, FPR, FNR, ERROR])
    errors.append(ERROR)
    
    score = roc_auc_score(yvl,lr.predict(xvl))
    print('ROC AUC score:',score)
    cv_score.append(score)
    
    i = i + 1
print('\nCv',cv_score,'\nMean CV Score',np.mean(cv_score))
all_vals.append(["Average of Errors", mean(errors), "", "", "", ""])
all_classifier.append(["Logistic Regression Classifier", mean(errors)])

header_names_result = ["Fold", "FP", "FN", "FPR", "FNR", "ERROR"]
res = pd.DataFrame(all_vals, columns=header_names_result)
res.set_index('Fold', inplace=True)
res
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

kf = StratifiedKFold(n_splits=5,shuffle=True,random_state=1)

cv_score =[]
all_vals = []
errors = []

i=1

for train_index,test_index in kf.split(X_train,y_train):
    print('{} of KFold {}'.format(i,kf.n_splits))
    xtr,xvl = X.loc[train_index],X.loc[test_index]
    ytr,yvl = y.loc[train_index],y.loc[test_index]
    
    #model
    dt = DecisionTreeClassifier().fit(xtr,ytr)
    tn, fp, fn, tp = confusion_matrix(yvl,dt.predict(xvl)).ravel()
    (FPR, FNR, ERROR) = display_results(tn, fp, fn, tp)
    all_vals.append([i, fp, fn, FPR, FNR, ERROR])
    errors.append(ERROR)
    
    score = roc_auc_score(yvl,dt.predict(xvl))
    print('ROC AUC score:',score)
    cv_score.append(score)
    
    i = i + 1
print('\nCv',cv_score,'\nMean CV Score',np.mean(cv_score))
all_vals.append(["Average of Errors", mean(errors), "", "", "", ""])
all_classifier.append(["Decision Tree Classifier", mean(errors)])

header_names_result = ["Fold", "FP", "FN", "FPR", "FNR", "ERROR"]
res = pd.DataFrame(all_vals, columns=header_names_result)
res.set_index('Fold', inplace=True)
res
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

kf = StratifiedKFold(n_splits=5,shuffle=True,random_state=1)

cv_score =[]
all_vals = []
errors = []

i=1

for train_index,test_index in kf.split(X_train,y_train):
    print('{} of KFold {}'.format(i,kf.n_splits))
    xtr,xvl = X.loc[train_index],X.loc[test_index]
    ytr,yvl = y.loc[train_index],y.loc[test_index]
    
    #model
    rfc = RandomForestClassifier(n_estimators=100).fit(xtr,ytr)
    tn, fp, fn, tp = confusion_matrix(yvl,rfc.predict(xvl)).ravel()
    (FPR, FNR, ERROR) = display_results(tn, fp, fn, tp)
    all_vals.append([i, fp, fn, FPR, FNR, ERROR])
    errors.append(ERROR)
    
    score = roc_auc_score(yvl,rfc.predict(xvl))
    print('ROC AUC score:',score)
    cv_score.append(score)
    
    i = i + 1
    
print('\nCv',cv_score,'\nMean CV Score',np.mean(cv_score))
all_vals.append(["Average of Errors", mean(errors), "", "", "", ""])
all_classifier.append(["Random Forest Classifier", mean(errors)])

header_names_result = ["Fold", "FP", "FN", "FPR", "FNR", "ERROR"]
res = pd.DataFrame(all_vals, columns=header_names_result)
res.set_index('Fold', inplace=True)
res
```

### K-Nearest Neighbors Classifer

```python
from sklearn.neighbors import KNeighborsClassifier

#Deciding value of K
error_rate = []

# Will take some time
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
kf = StratifiedKFold(n_splits=5,shuffle=True,random_state=1)

cv_score =[]
all_vals = []
errors = []

i=1

for train_index,test_index in kf.split(X_train,y_train):
    print('{} of KFold {}'.format(i,kf.n_splits))
    xtr,xvl = X.loc[train_index],X.loc[test_index]
    ytr,yvl = y.loc[train_index],y.loc[test_index]
    
    #model
    knn = KNeighborsClassifier(n_neighbors=5).fit(xtr,ytr)
    tn, fp, fn, tp = confusion_matrix(yvl,knn.predict(xvl)).ravel()
    (FPR, FNR, ERROR) = display_results(tn, fp, fn, tp)
    all_vals.append([i, fp, fn, FPR, FNR, ERROR])
    errors.append(ERROR)
    
    score = roc_auc_score(yvl,knn.predict(xvl))
    print('ROC AUC score:',score)
    cv_score.append(score)
    
    i = i + 1
    
print('\nCv',cv_score,'\nMean CV Score',np.mean(cv_score))
all_vals.append(["Average of Errors", mean(errors), "", "", "", ""])
all_classifier.append(["KNN Classifier", mean(errors)])

header_names_result = ["Fold", "FP", "FN", "FPR", "FNR", "ERROR"]
res = pd.DataFrame(all_vals, columns=header_names_result)
res.set_index('Fold', inplace=True)
res
```
